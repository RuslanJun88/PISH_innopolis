```python
import pandas as pd #подключение библиотеки pandas дял работы  сбазами данных./+
from sklearn.preprocessing import normalize
```

```python
data=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Files for lessons/world-data-2023.csv') #чтение базы данных в формате csv из google_drive./+
data                                                                                              #вывод данных
```

```python
tab=pd.DataFrame(data[['Country','Density\n(P/Km2)','Agricultural Land( %)']].iloc[[0,1,2,3,4]]) #Создание таблицы 1 для объединения
tab
```

```python
tab2=pd.DataFrame(data[['Abbreviation','Density\n(P/Km2)','Agricultural Land( %)']].iloc[[2,3,4,5,6]]) #Создание таблицы 2 для объединения
tab2
```

```python
# inner Merge: Возвращает только те строки, которые имеют совпадающие значения в ключевых
#столбцах и в левом, и в правом датафрейме. То есть, остаются только строки с общими значениями ключевых столбцов.

df1 = pd.DataFrame(tab)
df2 = pd.DataFrame(tab2)
result_m = pd.merge(df1, df2, on='Density\n(P/Km2)', how='inner')
result_m
```

```python
# Left Merge: возвращает все строки из левого (первого) датафрейма и только те строки из правого (второго) датафрейма,
# которые имеют совпадающие значения в ключевых столбцах. Если в правом датафрейме нет совпадающих значений,
# в результирующем датафрейме для них будут NaN.
df1 = pd.DataFrame(tab)
df2 = pd.DataFrame(tab2)
result = pd.merge(df1, df2, on='Density\n(P/Km2)', how='left')
result
```

```python
# Right Merge: возвращает все строки из правого (второго) датафрейма и только те строки из левого
# (первого) датафрейма, которые имеют совпадающие значения в ключевых столбцах.
df1 = pd.DataFrame(tab)
df2 = pd.DataFrame(tab2)
result = pd.merge(df1, df2, on='Density\n(P/Km2)', how='right')
result
```

```python
#Outer Join:
df1 = pd.DataFrame(tab)
df2 = pd.DataFrame(tab2)
result = pd.merge(df1, df2, on='Density\n(P/Km2)', how='outer')
result
```

```python
#Join по индексам
df1 = pd.DataFrame(tab)
df2 = pd.DataFrame(tab2)
result = pd.merge(df1, df2, left_index=True, right_index=True, how='inner')
result
```
